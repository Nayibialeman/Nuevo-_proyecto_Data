```python
# Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
# representación binaria. Debe recibir y devolver un valor de tipo entero.
# En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.

def NumeroBinario(numero):
    if not isinstance(numero, int) or numero < 0:  #si el numero no es int
        return None  # Si el número no es entero o es menor que cero, retorna None
    
    binario = ''  # Variable para almacenar el número binario resultante

    while numero > 0:  # Mientras el número sea mayor que cero
        residuo = numero % 2  # Calcula el residuo al dividir por 2
        binario = str(residuo) + binario  # Agrega el residuo a la izquierda del número binario
        numero //= 2  # Actualiza el número dividiéndolo por 2 (parte entera)

    if not binario:  # Si el número original era 0, su representación binaria es '0'
        return 0
    else:
        return int(binario)
```
